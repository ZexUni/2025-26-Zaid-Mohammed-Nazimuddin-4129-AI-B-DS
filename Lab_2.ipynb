```python
n=int(input("Enter number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element{i+1}:")))
key=int(input("Enter element to search:"))
found=False
for i in range(n):
  if arr[i]==key:
    print(f"Element {key} found at position {i+1}")
    found=True
if found is not True:
  print(f"{key} is not found")
```

```python
n=int(input("Enter number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}:")))
arr.sort()
print("\n Sorted list:",arr)
key=int(input(f"Enter element to search:"))
low=0
high=n-1
found=False
while low<=high:
  mid=(low+high)//2
  if arr[mid]==key:
    print(f"Element {key} found at position {mid+1}")
    found=True
    break
  elif key<mid:
    high=mid-1
  else:
    low=mid+1
if not found:
    print("Element not found")
```

```python
import pandas as pd
data={
    "Department":["IT","IT","HR","HR","Finance"],
    "Name": ["A","B","C","D","E"],
    "Marks": [80, 85, 70, 95, 88]
    }
df=pd.DataFrame(data)
print("Original Data:")
print(df)
print("\nAverage Marks by Department:")
print(df.groupby("Department")["Marks"].mean())
print("\nTotal Marks by Department:")
print(df.groupby("Department")["Marks"].sum())
print("\nStudent Count of Department:")
print(df.groupby("Department")["Marks"].count())
```

```python
import pandas as pd
data={
    "Name":["A","B","C"],
    "Age":[22,21,23],
    "Marks":[85,90,95]
}
df=pd.DataFrame(data)
print("\nOriginal Data:")
print(df)
df_sorted=df.sort_values(by="Marks")
df_sorted_desc=df.sort_values(by="Marks",ascending=False)
print("\n After rearranging rows by Marks (descending):")
print(df_sorted_desc)
df_rearranged_cols=df[["Marks","Name","Age"]]
print("\n After rearranging columns:")
print(df_rearranged_cols)
```

```python
import pandas as pd
data={
    "Name":["A","B","C","D"],
    "Marks":[85,92,78,92],
    }
df=pd.DataFrame(data)
print("Original Data:")
print(df)
df["Rank"]=df["Marks"].rank(ascending=False)
print("\nAfter Ranking:")
print(df)
```
