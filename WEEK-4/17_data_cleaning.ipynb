```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, None, 35, 40, 28],
    'Salary': [50000, 60000, None, 70000, 62000],
    'Allowances': [2000, None, 2000, 2000, 2500]
}

df = pd.DataFrame(data)

# (a) Show missing values (True/False)
print("Missing Values (True/False):")
print(df.isnull())

# (b) Sum of missing values
print("\nMissing Values Count per Column:")
print(df.isnull().sum())

print("\nTotal Missing Values in Dataset:")
print(df.isnull().sum().sum())
```
